```python
def discount(Price):
    if Price >= 100 and Price<200:
        return 5
    elif Price >=200 and Price<500:
        return 10
    elif Price >=500 and Price<=1000:
        return 15
    elif Price>1000:
        return 30
    else:
        return 0
```
