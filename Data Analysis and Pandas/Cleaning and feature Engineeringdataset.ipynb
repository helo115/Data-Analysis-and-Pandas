## Feature Engineering

```python
def discount(price):
    if 1500<price<=5000:
        return "5%"
    elif 5000<price<=50000:
        return "10%"
    elif 50000<price<=100000:
        return "20%"
    else:
        return "0%"
```

```python
dataset["Discount"]=dataset["Price"].apply(discount)
```

```python
dataset
```

```python
def discount_price(discount,price):
    if discount=="5%":
        return price*5/100
    elif discount=="10%":
        return price*10/100
    elif discount == "15%":
        return price*15/100
    elif discount == "20%":
        return price*20/100
    else:
        return 0 
```

```python
dataset["Discount_Price"]=dataset["Price"]*dataset["Discount"].str.rstrip("%").astype(int)/100
```

```python
dataset
```

```python
dataset["Final Price"]=dataset["Price"]-dataset["Discount_Price"]
```

```python
dataset
```

```python
dataset["Date"]=dataset["Date"].str.replace("/","-")
```
