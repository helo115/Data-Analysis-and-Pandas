```python

# Sample dataset
data = {
    "Customer_Name": ["Ali","Mariam","Usman","Tom","Sana","Ali","Mariam","Usman","Tom","Sana"],
    "City": ["Karachi","Lahore","Multan","Islamabad","Karachi","Lahore","Multan","Islamabad","Karachi","Lahore"],
    "Product": ["Laptop","Mobile","Keyboard","Charger","Smart Watch","Laptop","Mobile","Keyboard","Charger","Smart Watch"],
    "Quantity": [2,1,3,1,2,1,2,1,3,1],
    "Price": [1200,300,30,10,80,1200,300,30,10,80],
}
```

```python
import pandas as pd
import numpy as np 
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Total Price"]=df["Price"]*df["Quantity"]
```
