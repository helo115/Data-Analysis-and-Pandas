```python
def Unit(Unit):
    if 0>Unit==-1:
        return 1
    elif 0<Unit==-2:
        return 2
    elif 0<Unit==-3:
        return 3
    else:
        return Unit
```

```python
df["Units_Sold"].min()
```

```python
df["Units_Sold"]=df["Units_Sold"].apply(Unit)
```

```python
df
```

```python
df["Units_Sold"]=df["Units_Sold"].fillna(df["Units_Sold"].mean())
```

```python
df
```

```python
df["Discount_Applied"]=df["Discount_Applied"].fillna(0)
```

```python
df
```

```python
#Revenue = Units_Sold × MRP × (1 – Discount_Applied)
df["Revenue"]=df["Units_Sold"]*df["MRP"]*(1-df["Discount_Applied"])
```

# Feature Engineering

# Calculating Profit on 20% Margin

```python
df["Profit"]=df["Revenue"]*0.20
```

```python
df
```

```python
def Size(size):
    if size=="6" or size=="7":
        return "S"
    elif size=="8" or size=="9":
        return "M"
    elif size=="10" or size=="11":
        return "L"
    elif size=="12" or size=="13" or size=="14" or size=="15":
        return "XL"
    else:
        return size
```

```python
df["Size"]=df["Size"].apply(Size)
```

```python
from datetime import datetime

def clean_date(date_str):
    formats = ["%Y/%m/%d", "%Y-%m-%d", "%d-%m-%Y", "%d/%m/%Y"]
    for fmt in formats:
        try:
            return datetime.strptime(date_str, fmt)
        except:
            continue
    return pd.NaT  # fallback if none match

df["Order_Date"] = df["Order_Date"].astype(str).apply(clean_date)
```
