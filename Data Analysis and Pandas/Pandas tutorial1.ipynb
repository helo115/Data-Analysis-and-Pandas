```python
def categories(BMI):
    if BMI <18.5:
        return "UNDER WEIGHT"
    elif 18.5<= BMI < 25:
        return "Normal"
    elif 25<=BMI<30:
        return "Overweight"
    elif BMI==30:
        return "Obesity"
    elif 30<BMI<35:
        return "Obesity Class I"
    else:
        return "Obesity class II"
```

```python
df_ext
```

```python
df_ext["Categories Range"]=df_ext["BMI"].apply(categories)
```

```python
df_ext
```

```python
df_ext[df_ext["BMI"]>29.9]
```

```python
bmi_counts=df_ext["Categories Range"].value_counts()
```

```python
bmi_counts
```

```python
import matplotlib.pyplot as plt
```

```python
bmi_counts.index
```

```python
bmi_counts.values
```

```python
plt.bar(bmi_counts.index,bmi_counts.values,color=["RED","orange","SkyBlue","Yellow","Brown"])
plt.title("BMI CATEGORIES RANGE CHART")
plt.xlabel("Categories",labelpad=10)
plt.ylabel("Number of People",labelpad=8)
plt.xticks(rotation=40)
```

```python
order=["UNDER WEIGHT","Normal","Obesity Class I","Overweight","Obesity class II"]
```

```python
NEW_ORDER=df_ext["Categories Range"].value_counts().reindex(order)
```
