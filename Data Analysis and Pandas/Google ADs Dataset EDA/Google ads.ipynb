```python
from datetime import datetime

def parse_date(x):
    for fmt in ("%d-%m-%Y", "%Y-%m-%d"):
        try:
            return datetime.strptime(x, fmt)
        except:
            continue
    return pd.NaT
```

```python
df["Ad_Date"]=df["Ad_Date"].apply(parse_date)
```

```python
df=df.dropna(subset=["Ad_Date"])
```
